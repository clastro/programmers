```python
PID_list = []
white_year_set = set()
pink_year_set = set()
blue_year_set = set()
for i in range(len(jobhistory)):
    if(i == 0):
        pass
    else:
        current = int(jobhistory.loc[i,["PID"]])
        prev = int(jobhistory.loc[i-1,["PID"]])
        if(current == prev):
            job_type = int(jobhistory.loc[i,["job_type"]])
            
            for j in range(int(jobhistory.loc[i,["begin_year"]]),int(jobhistory.loc[i,["finish_year"]])+1):                
                if(job_type == 1):
                    white_year_set.add(j)
                elif(job_type == 2):
                    pink_year_set.add(j)
                elif(job_type == 3):
                    blue_year_set.add(j)
        else:
            PID_list.append([len(white_year_set),len(pink_year_set),len(blue_year_set)])
            white_year_set = set()
            pink_year_set = set()
            blue_year_set = set()
            job_type = int(jobhistory.loc[i,["job_type"]])
            for j in range(int(jobhistory.loc[i,["begin_year"]]),int(jobhistory.loc[i,["finish_year"]])+1):                
                if(job_type == 1):
                    white_year_set.add(j)
                elif(job_type == 2):
                    pink_year_set.add(j)
                elif(job_type == 3):
                    blue_year_set.add(j)
            
                
```
