```python
PID_list = [] #리스트를 담을 변수
white_year_set = set() # 연도가 들어갈 화이트 컬러 직업의 집합 생성
pink_year_set = set() # 연도가 들어갈 핑크 컬러 직업의 집합 생성
blue_year_set = set() # 연도가 들어갈 블루 컬러 직업의 집합 생성
for i in range(len(jobhistory)): #jobhistory의 데이터만큼 반복문 시작
    if(i == 0): #시작일 경우 pass
        pass
    else: #두번째 데이터부터 쿼리
        current = int(jobhistory.loc[i,["PID"]]) #현재 PID값
        prev = int(jobhistory.loc[i-1,["PID"]]) #이전 PID값
        if(current == prev): #PID 값이 계속 같을 경우
            job_type = int(jobhistory.loc[i,["job_type"]])
            
            for j in range(int(jobhistory.loc[i,["begin_year"]]),int(jobhistory.loc[i,["finish_year"]])): #처음 연도와 끝나는 연도 loop                
                if(job_type == 1): #job_type이 1, white일 경우
                    white_year_set.add(j) #화이트 컬러 집합에 연도를 추가한다
                elif(job_type == 2):
                    pink_year_set.add(j)
                elif(job_type == 3):
                    blue_year_set.add(j)
        else: #PID 값이 다를 경우
            PID_list.append([len(white_year_set),len(pink_year_set),len(blue_year_set)]) #기존 PID 값의 연도 집합 데이터를 리스트에 저장
            white_year_set = set() #white 다시 생성
            pink_year_set = set() #pink 다시 생성
            blue_year_set = set() #blue 다시 생성
            #PID가 바뀔 때도 위의 반복문과 똑같이 처음연도와 끝나는 연도를 추가해준다
            job_type = int(jobhistory.loc[i,["job_type"]])
            for j in range(int(jobhistory.loc[i,["begin_year"]]),int(jobhistory.loc[i,["finish_year"]])):                
                if(job_type == 1):
                    white_year_set.add(j)
                elif(job_type == 2):
                    pink_year_set.add(j)
                elif(job_type == 3):
                    blue_year_set.add(j)
            
                
```
